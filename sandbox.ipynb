```python
import numpy as np
```

```python
wd = np.arange(355, 370)
```

```python
sin_sum = sum(np.sin(np.radians(wd)))
cos_sum = sum(np.cos(np.radians(wd)))
```

```python
sin_sum
```

```python
cos_sum
```

```python
np.degrees(np.arctan(sin_sum/cos_sum))
```

```python
np.degrees(np.arctan2(sin_sum,cos_sum))
```
